```python
import spacy
nlp= spacy.load("en_core_web_lg")
```

```python
doc=nlp("church mosque muslim christian traditional occult web")
for token in doc:
    print(token.text, "Vector: ", token.has_vector, "OOV: ", token.is_oov)
```

```python
test_token = nlp("religion")
```

```python
for token in doc:
    print(f"{token.text} <-> {test_token}", token.similarity(test_token))
```

```python
df2['sentiment_binary_rep'] = df2['sentiment'].map({'positive':1, 'negative':0})
```

```python
df2.head(1700)
```

```python
df2['review_vector']=df2['review'].apply(lambda x: nlp(x).vector)
```

```python
df2.head(1730)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df2.review_vector.values,
    df2.sentiment_binary_rep,
    test_size = 0.2,
    random_state=2022 
)
```

```python
X_train
```

```python
import numpy as np

X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
X_train_2d
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_train_embd = scaler.fit_transform(X_train_2d)
scaled_test_embd = scaler.transform(X_test_2d)

clf = MultinomialNB()
clf.fit(scaled_train_embd, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = clf.predict(scaled_test_embd)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')
clf.fit(X_train_2d, y_train)
y_pred = clf.predict(X_test_2d)
print(classification_report(y_test, y_pred))
```
